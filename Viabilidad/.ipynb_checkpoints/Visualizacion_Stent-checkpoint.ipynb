```python
import pyvista as pv

# Cargar el archivo STL
archivo_stl = r"C:\Users\julia\Desktop\ING_SALUD\TFG\stent_1.stl"
malha = pv.read(archivo_stl)

# Crear una ventana interactiva
plotter = pv.Plotter()
plotter.add_mesh(malha, color="lightblue", show_edges=True)
plotter.show()
```

```python
import vtk

# Cargar el archivo STL
reader = vtk.vtkSTLReader()
reader.SetFileName(r"C:\Users\julia\Desktop\ING_SALUD\TFG\stent_1.stl")

# Crear un mapeador
mapper = vtk.vtkPolyDataMapper()
mapper.SetInputConnection(reader.GetOutputPort())

# Crear un actor
actor = vtk.vtkActor()
actor.SetMapper(mapper)

# Crear un renderizador
renderer = vtk.vtkRenderer()
renderer.AddActor(actor)

# Crear una ventana de renderizado
render_window = vtk.vtkRenderWindow()
render_window.AddRenderer(renderer)

# Crear un interactor
render_window_interactor = vtk.vtkRenderWindowInteractor()
render_window_interactor.SetRenderWindow(render_window)

# Renderizar y mostrar
render_window.Render()
render_window_interactor.Start()
```
