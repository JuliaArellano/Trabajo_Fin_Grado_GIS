Expansión del stent según la temperatura 

Se emplea un modelo fenomenológico tipo sigmoide para describir la fracción de transformación de fase, acoplado con una ley de mezclas para interpolar propiedades térmicas, dentro del marco de la mecánica del continuo.



## *1. Fundamentos de la Transformación de Fases en Nitinol*

El *nitinol* es una aleación con memoria de forma que puede existir en dos fases cristalinas principales:

- *Martensita (baja temperatura, baja rigidez, deformable)*  
- *Austenita (alta temperatura, más rígida, fase original del stent)*

La transformación entre estas fases no ocurre instantáneamente sino en un *rango de temperatura*, lo que da lugar a una fracción de transformación que varía suavemente con la temperatura.

### *1.1. Temperaturas Características de la Transformación*

El cambio de fase en nitinol ocurre dentro de un intervalo térmico determinado por cuatro temperaturas clave:

- \( M_s \) – Temperatura de inicio de la transformación *a martensita*  
- \( M_f \) – Temperatura final de la transformación *a martensita*  
- \( A_s \) – Temperatura de inicio de la transformación *a austenita*  
- \( A_f \) – Temperatura final de la transformación *a austenita*  

Cuando se calienta el material, la martensita se transforma en austenita en el intervalo \( [A_s, A_f] \).  
Cuando se enfría, la austenita se transforma en martensita en el intervalo \( [M_s, M_f] \).

## *2. Expansión Térmica Basada en la Transformación de Fase*
El nitinol tiene *dos coeficientes de expansión térmica* distintos:  
- *Martensita*: \( \alpha_m \) (pequeño, aproximadamente \( 1 \times 10^{-6} \, \text{°C}^{-1} \))  
- *Austenita*: \( \alpha_a \) (mayor, aproximadamente \( 6 \times 10^{-6} \, \text{°C}^{-1} \))  

Como el material está compuesto por una mezcla de estas dos fases según la fracción \( \xi(T) \), el *coeficiente de expansión térmica efectivo* del nitinol se define como:

\xi(T) = \frac{1}{2} \left[ 1 + \tanh\left( \frac{T - M_s}{\Delta T} \right) \right]

## *3. Expresión Final para la Expansión Radial*
Dado un stent de radio inicial \( r_0 \), el cambio de radio en función de la temperatura es:

\[
\Delta r = r_0 \left[ \alpha_m (1 - \xi) + \alpha_a \xi \right] \cdot \Delta T
\]

donde:
- \( \Delta r \) es la expansión radial del stent.
- \( \xi(T) \) es la fracción de austenita en función de la temperatura.
- \( \Delta T = T - T_0 \), donde \( T_0 \) es la temperatura inicial de referencia.

### Aproximación utilizada
| **Fórmula** | **Nombre / Tipo de Aproximación** | **Descripción** |
|-------------|------------------------------------|------------------|
| \( \xi(T) = \frac{1}{2} \left[ 1 + \tanh\left( \frac{T - M_s}{\Delta T} \right) \right] \) | Función sigmoide térmica / Modelo fenomenológico | Modela la fracción de austenita en función de la temperatura de forma suave y continua. |
| \( \alpha_{\text{eff}}(T) = \alpha_m (1 - \xi) + \alpha_a \xi \) | Ley de mezclas (interpolación de fases) | Calcula el coeficiente de expansión térmica efectivo según la fracción de fase. |
| \( \Delta r = r_0 \cdot \alpha_{\text{eff}}(T) \cdot \Delta T \) | Expansión térmica no lineal acoplada | Extiende la ley clásica de dilatación térmica al caso de materiales con cambio de fase dependiente de temperatura. |


```python
import numpy as np
import matplotlib.pyplot as plt

# Temperaturas de transformación
M_s = 20  # Inicio martensita (°C)
M_f = 30  # Fin martensita (°C)
A_s = 30  # Inicio austenita (°C)
A_f = 40  # Fin austenita (°C)
T_max = 37  # Temperatura donde el radio debe ser 3.5 mm

# Rango de temperaturas para la simulación
temperaturas = np.linspace(20, 60, 200)

#Función para calcular la fracción de transformación de fase
def fraccion_transformacion(T, T_inicio, T_fin):
    delta_T = T_fin - T_inicio
    return 0.5 * (1 + np.tanh((T - T_inicio) / delta_T))

# Fracción de transformación de martensita a austenita
austenita_frac = fraccion_transformacion(temperaturas, M_s, A_f)
# Parámetros del stent
r_0 = 0.00  # Radio inicial en metros (1.5 mm)
r_f = 0.0035  # Radio final en metros (3.5 mm)
alpha_m = 0  # Coeficiente de expansión térmica para martensita
alpha_a = 6e-6  # Coeficiente de expansión térmica para austenita

# Índice donde la temperatura es 37°C
idx_37 = np.argmax(temperaturas >= T_max)

# Calcular expansión por transformación de fase antes de escalar
expansion_transformacion = austenita_frac

# Ajustar la transformación para que r_f se alcance exactamente en T_max
scaling_factor = (r_f - r_0) / expansion_transformacion[idx_37]
expansion_transformacion *= scaling_factor

# Cambio de radio por expansión térmica limitado hasta T_max
expansion_termica = (alpha_m * (1 - austenita_frac) + alpha_a * austenita_frac) * (np.minimum(temperaturas, T_max) - 20)

# Expansión total sin contribuciones adicionales después de T_max
expansion_radial = expansion_transformacion + expansion_termica

# Ajustar el radio máximo a exactamente 3.5 mm
radio_stent = r_0 + expansion_radial
radio_stent *= r_f / radio_stent[idx_37]

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(temperaturas, radio_stent, label="Radio del Stent", color="blue")
plt.axvline(M_s, linestyle='--', color='orange', label="Inicio martensita (M_s)")
plt.axvline(M_f, linestyle='--', color='brown', label="Fin martensita (M_f)")
plt.axvline(A_s, linestyle='--', color='r', label="Inicio austenita (A_s)")
plt.axvline(A_f, linestyle='--', color='g', label="Fin austenita (A_f)")
plt.axvline(T_max, linestyle='--', color='purple', label="Expansión Máxima (37°C)")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Radio del Stent (m)")
plt.title("Expansión del Stent en función de la Temperatura")
plt.grid(True)
plt.legend()
plt.show()
```
